# Current Steering DAC Unit Cell Design

The following details the design procedure for the design of the unit cell used in a 10-bit current steering DAC.

The topology uses 256 segmented cells employing a thermometer code to form 8-bits of resolution. Further partial segmentation is achieved with half and quarter current cells to provide the final 2-bits.

The design is made the Skywater 130 nm process using only 1.8 V devices.

## Current Source Design
Firstly the current source will be designed.

#### Matching

It has been shown that for a random change, $\sigma_I$, in the unit current, $I_u$,:  
$$INL_{max} = \frac{\sigma_I}{2 I_u} \sqrt{2^N} LSB$$

The goal is to size and bias the current source such that $INL_{max} < 1 LSB$, therefore, $\frac{\sigma_I}{2 I_u} \sqrt{2^N} < 1$.

This allows the matching goal to be defined as:
$$ \frac{\sigma_I}{I_u} < \frac{2}{\sqrt{2^N}}  $$


The Skywater 130 nm has no publised mismatch data, however let's use the conservative rule of thumb for threshold voltage mismatch of $ A_{VTH} = 10~mV \cdot \mu m$ can be used. Where $\sigma_{VTH} = A_{VTH} / \sqrt{WL}$. Using square law approximations we can state that:

$$ \frac{\sigma_I}{I_u} = \frac{2 \sigma_{VTH}}{V_{GS}-V_{TH}} = \frac{2 A_{VTH}}{\sqrt{WL} \cdot (V_{GS}-V_{TH})} $$

Finally, we can re-arrange to place the parameters that we use to design on one side:

$$ \sqrt{WL} \cdot (V_{GS}-V_{TH}) = 2 A_{VTH} \frac{I_u}{\sigma_I} $$

$$ \sqrt{WL} \cdot (V_{GS}-V_{TH}) = A_{VTH} \sqrt{2^N} $$

We're targetting 3 sigma ($3 \cdot \sigma$) and a 10-bit design so

$$ \sqrt{WL} \cdot (V_{GS}-V_{TH}) > 960 ~~~~~~ [mV \cdot \mu m]  $$

```python
target = {'sqrt(wl).Vov'  : 0.96}
```

#### Integrated Noise

Let's plot the integrated noise for each channel length. We're using a bandwidth of 10 MHz.

We will inspect the $L = 8 \mu m$ as we know a large area will be required and high overdrive.
We can see a minimum in the integrated noise at $ \frac{g_m}{I_d} = 5$.

```python
conditions = {  'l'     :   query_obj['pfet_lvt'].get_parameter_values('l'),
                'vbs'   :   0.00,
                'vds'   :   0.40,
                'f_hi'  :   10e6}
query_obj['pfet_lvt'].plot(x='gm/id', y='integrated_noise', conditions=conditions, extra_plot_cmd='plt.ylim([1e-10,1e-7])')
```

#### Vdsat

Provisionally we'll select $ \frac{g_m}{I_d} = 5$ and now we inspect $ V_{dssat}$.

We can 

```python
selected = {'gm/id' : 5,
            'l'     : 8.00}

conditions = {  'l'     :   selected['l'],
                'vbs'   :   0.00,
                'vds'   :   0.40}
selected['vdsat'] = query_obj['pfet_lvt'].get_matching_value(original   = 'gm/id', 
                                                             matching   = 'vdsat', 
                                                             value      = selected['gm/id'], 
                                                             conditions = conditions)
print('Selected Vdsat = %.1f mV' % (selected['vdsat']*1000))
```

#### Calculate required area
Now that we have the selected overdrive voltage we can calculate the required area for matching:

```python
target['A'] = (target['sqrt(wl).Vov'] / selected['vdsat'])**2
target['w'] = target['A'] / selected['l']
print('Minimum gate width to meet matching = %0.2f um' % target['w'])
```

#### Calculate required current
Finally calculate the current required to meet the selected width, length and transconductance efficiency.

```python
selected['id/w'] = query_obj['pfet_lvt'].get_matching_value(original   = 'gm/id', 
                                                            matching   = 'id', 
                                                            value      = selected['gm/id'], 
                                                            conditions = conditions)
selected['w']  = 1
selected['id'] = selected['id/w'] * selected['w']
```

### Final results
The following is the designed unit cell

```python
print('Gate width / length         = %0.1f / %0.1f \t[um]'  % (selected['w'], selected['l']))
print('Drain current               = %0.1f  \t\t[uA]'  % (selected['id'] * 1e6))
print('Transconductance efficiency = %0.1f  \t\t[S/A]' % (selected['gm/id']))
print('Saturation voltage          = %0.1f  \t\t[mV]'  % (selected['vdsat']*1e3))
```

```python
print('Total static current consumption = %0.1f \t[uA]' % (258 * selected['id'] * 1e6))
print('Total static power consumption   = %0.1f \t[uW]' % (1.8 * 258 * selected['id'] * 1e6))
```
